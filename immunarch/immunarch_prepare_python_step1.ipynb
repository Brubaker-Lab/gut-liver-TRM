```python
def revert_format_TRBV_column(df, column_name):
    """
    Reverts the format of the 'TRBV' column entries by removing leading zeros from the indices.
    """
    def revert_format_TRBV_value(value):
        # This function uses regular expressions to find and remove leading zeros in the numbers
        return re.sub(r'(\D)0(\d)', r'\g<1>\2', value)

    df[column_name] = df[column_name].apply(revert_format_TRBV_value)
    return df
```
